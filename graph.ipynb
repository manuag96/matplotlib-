```python
import matplotlib.pyplot as mat
x=[0,8]
y=[5,9]

mat.xlabel("time")         # to give label to x axis
mat.ylabel("speed")        #to give label to y axis
mat.plot(x,y)
mat.show()
```

```python
import matplotlib.pyplot as mat
x=[0,8]
y=[5,9]

x1=[0,9]
y1=[7,5]
mat.xlabel("time")                # to give label to x axis
mat.ylabel("speed")               #to give label to y axis

mat.plot(x,y,label="bicycle")  
mat.plot(x1,y1,label="car")
mat.legend()                     # to give label to graph
mat.show()
```

```python
import matplotlib.pyplot as mat
x=[0,8]
y=[5,9]

x1=[0,9]
y1=[7,5]
mat.xlabel("time")         # to give label to x axis
mat.ylabel("speed")        #to give label to y axis

mat.plot(x,y,label="bicycle")  
mat.plot(x1,y1,label="car")
mat.legend()               # to give label to graph
mat.grid()                 # forms grid in the graph 
mat.show()
```

```python
import matplotlib.pyplot as mat
x=[0,8]
y=[5,9]

x1=[0,9]
y1=[7,5]
mat.xlabel("time",)         # to give label to x axis
mat.ylabel("speed")                      #to give label to y axis

mat.plot(x,y,label="bicycle")  
mat.plot(x1,y1,label="car")
mat.legend()                             # to give label to graph
mat.grid( color='black')                 # forms grid in the graph and gives colour 
mat.show()
```

```python
import matplotlib.pyplot as mat
x=[4,8]
y=[5,9]

x1=[0,6]
y1=[7,5]
mat.xlabel("time",)                      # to give label to x axis
mat.ylabel("speed")                      #to give label to y axis

mat.bar(x,y,label="bicycle")  
mat.bar(x1,y1,label="car")
mat.legend()                             # to give label to graph
mat.grid( color='black')                 # forms grid in the graph and gives colour 
mat.show()
```

```python
import matplotlib.pyplot as mat
x=[0,8,7,6]
y=[5,9,6,2]

x1=[0,9,3,1]
y1=[7,5,6,3]
mat.xlabel("time",)                      # to give label to x axis
mat.ylabel("speed")                      #to give label to y axis

mat.scatter(x,y,label="bicycle")  
mat.scatter(x1,y1,label="car")          
mat.legend()                             # to give label to graph
mat.grid( color='black')                 # forms grid in the graph and gives colour 
mat.show()
```

```python
import matplotlib.pyplot as mat
x=[2000,2001,2004,2005,2006,2007]
y=[7,8,9,10,15,6]
mat.xlabel("year")
mat.ylabel("population")
mat.scatter(x,y,marker='x',color='r')     #to give marker as 'x' instead of a dot

mat.grid(color='green')
mat.show()
```
